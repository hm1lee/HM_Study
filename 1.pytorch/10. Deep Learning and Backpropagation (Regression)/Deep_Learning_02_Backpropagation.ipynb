## Backpropagation

**Our Objective**

* 주어진 데이터에 대해서 출력 값을 모사하는 함수를 찾고 싶다.

* Loss 값을 최소로 하는 Loss Function의 입력 값 θ을 찾자.

* Gradient Descent를 통해서 현재 θt에서 더 나은 θt+1로 나아가자. (t는 iteration의 횟수)

* DNN은 파라미터만 많을 뿐 기존과 같은 과정을 거친다.

**Without Backpropagation..**

- Backpropagation으로 미분 가능
- Loss값을 학습 파라미터로 미분 해줘야 한다.

**Summary**

* Gradient Descent를 통해서 loss를 최소화하는 과정은 신경망 파라미터를 찾는 과정이다. 
    * 하지만 loss를 각 레이어의 파라미터로 그냥 미분하는 과정은 매우매우 **비효율적**
    * 따라서 back-propagation을 통해서 반복되는 미분 과정을 효율적으로 만들 수 있다.

* Pytorch, Tensorflow 등의 프레임워크는 AutoGrad와 같은 기능을 통해서 fee-forward 작업에 대해서 자동 미분을 수행해준다.
