## Stochastic Gradient Descent (SGD)

- 전체 샘플의 loss에 대한 gradient descent가 아닌, 일부 샘플(k = batch size)의 loss에 대한 gradient descent.

- 1 epoch (전체 데이터셋을 활용한 학습)당 파라미터 update 횟수 증가

- batch_size가 작아질수록 gradient가 실제 gradient와 달라질 것

- 요즘 학계 추세 : 큰 batch size를 가져가려고 함 
   * gpu 사용으로 병렬 연산이 가능하여, 큰 배치사이즈로 인한 비용이 커버되기 때문에
   * 매우 큰 배치사이즈의 경우 오히려 성능을 악화시킬 수 있음

**Stochastic Gradient Descent vs Gradient Descent**

- 각 미니배치의 gradient는 전체 데이터의 gradient와 다를 것 (noisy)

- 하지만 미니배치의 gradient 평균은 전체 데이터의 gradient와 같을 것 

- Batch size가 작을 수록 variance는 더욱 커질 것

**Batch SIze**

- Batch size가 작으면 : loacl minima를 탈출할 수도 있음

- Batch size가 크면 : 상대적으로 더 빠르게 수렴할 수 있음

**IF Batch Size equals to Dataset Size**

- 데이터셋의 크기가 매우 크다면, 데이터셋의 분포는 실제 모집단의 분포와 비슷해질 것 

- gradient의 방향은 실제 전체 데이터셋을 활용한 gradient의 방향과 비슷할 것

- 한번의 파라미터 업데이트를 위해서 더 많은 계산이 필요함
