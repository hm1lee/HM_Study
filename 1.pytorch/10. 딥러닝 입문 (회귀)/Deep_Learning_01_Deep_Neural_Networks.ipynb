## Deep Neural Networks

**1.Linear & Logistic Regression**

**Motivation**
* 데이터의 선형 관계를 분석 및 예측
* 세상에는 차원이 높고, 복잡한 비선형 데이터들이 많음 
  * 고차원 데이터들은 plot으로 확인하기도 어려움
* 만약 linear layer를 깊게 쌓으면 안될까?

image.png

**How?**
* linear를 깊게 쌓아봤자, linear이다.
* 우리가 해결하고 싶은 문제는 **non-linear**
* 이전에 Logistic Regression에서 **Sigmoid**함수를 통해 매우 작은 값은 0에 수렴하고, 매우 큰 값은 1에 수렴하도록 하였음
image.png

* 레이어 사이에 비선형 활성 함수를 넣어서 깊게 쌓아보자! (꼭 sigmoid일 필욘 없다!)
image.png

**Network Capacity**

깊이(depth)와 너비(width)를 넓게 할수록 신경망의 표현력은 좋아진다! 
   * 복잡한 형태의 함수를 배울 수 있는 능력이 생긴다.

   image.png

**Summary**

* 많은 문제들이 굉장히 높고 복잡한 차원의 데이터를 다룸 (이미지, 텍스트, 음성 ...)

* 기존의 선형 데이터 분석 기법으로는 한계가 명확함

* Non-linear 함수를 레이어 사이에 넣어주어 비선형 모델을 만들 수 있음

* 네트워크의 파라미터들을 gradient descent를 통해서 학습하면, target 함수를 모사(approximate)할 수 있음
