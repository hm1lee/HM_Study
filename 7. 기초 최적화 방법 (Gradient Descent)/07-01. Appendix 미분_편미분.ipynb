## Appendix : Brief Introduction to Derivative
### (부록 : 파생 모델에 대한 간략한 소개)

1. 기울기 : 함수의 두 입력 값에 대한 출력 값의 변화량의 비율



2. 극한(무한소)과 미분
- 두 점이 한없이 가까워질때





3. 뉴턴 vs 라이프니츠

- 뉴턴 미분법 (변수가 하나일 때 편리)
- 라이프니츠 미분법 (변수가 두 개 이상일 때 편리) -> 앞으로 자주 사용하기 때문에 익숙해질 필요가 있다.
   


4. 다변수 함수
- 여러 개의 변수를 입력으로 받는 함수 
- z = f(x, y)

5. 편미분

- 다변수 x와 y를 입력으로 받는 함수 f를 x로 미분할 경우, 하나의 변수만 남겨놓고 나머지를 상수 취급 하는 방법)
- 예) 함수 f를 x로 미분 

6. 함수의 입출력 형태
- 함수의 입력이 벡터인 경우
- 함수의 입력이 행렬인 경우
- 함수의 출력이 백터인 경우
- 함수의 출력이 행렬인 경우
- 입력과 출력이 벡터인 함수
