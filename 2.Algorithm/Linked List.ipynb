1. 연결 리스트 구현하기



```python
class ListNode:
    def __init__(self,val):
        self.val  = val
        self.next = None
```



ListNode(1)를 하게 되면, 위와 같이 val값이 할당되며, next는 None으로 지정된다.

```python
head_node = ListNode(1)
```

```python
head_node.next = ListNode(2)
head_node.next.next = ListNode(3)
head_node.next.next.next = ListNode(4)
```

2. 연결리스트 value 출력하기



```python
# iterative한 방법 (하나씩 탐색)
def printNodes(node:ListNode):
    crnt_node = node
    while crnt_node is not None :
        print(crnt_node.val, end = ' ')
        crnt_node = crnt_node.next

printNodes(head_node)
```

```python
# recursive한 방법
def printNodesRecur(node:ListNode):
    print(node.val, end=' ')
    if node.next is not None :
        printNodesRecur(node.next)
    
printNodesRecur(head_node)
```

3. 연결리스트 Class로 만들기
- add at Head
- add at back / after
- Find
- Delete

```python
# Add at Head

class ListNode:
    def __init__(self,val):
        self.val = val
        self.next = None


def printNodes(node:ListNode):
    crnt_node = node
    while crnt_node is not None:
        print(crnt_node.val, end=' ')
        crnt_node = crnt_node.next
```

```python
class SLinkedList : 
    def __init__ (self):
        self.head = None
    
    def addAtHead(self, val):
        node = ListNode(val)
        node.next = self.head  #self.head가 가지고 있던 None을 node.next에 넣는다 = None이 된다
        self.head = node  #self.head의 빈자리가 node를 가르킨다
```

```python
slist = SLinkedList()
slist.addAtHead(1)
slist.addAtHead(2)
printNodes(slist.head)
```
